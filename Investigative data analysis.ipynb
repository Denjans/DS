Изучите зависимость цены от того, на каком этаже расположена квартира: первом, последнем или другом. Также изучите зависимость от даты размещения: дня недели, месяца и года.

```python
def display_corr(data, corr1, corr2, kind='scatter', alpha=0.05, bins=10, gridsize=10):
    print(f'Коэф Пирсона {corr1} от {corr2}:', data[corr1].corr(data[corr2]))
    if kind=='scatter':
        print(data.plot(x=corr1, y=corr2, kind=kind, alpha=alpha))
    elif kind=='bar':
        print(data.plot(x=corr1, y=corr2, kind=kind))
    elif kind=='hist':
        print(data.plot(x=corr1, y=corr2, kind=kind, bins=bins))
    elif kind=='hexbin':
        print(data.plot(x=corr1, y=corr2, kind=kind, gridsize=gridsize)) 
    elif kind=='line':
        print(data.plot(x=corr1, y=corr2, kind=kind))
    else:
        print(data.groupby(corr1).agg({corr2: 'median'}).plot())
```

```python
#усредним данные по стоимости м2 и сгруппируем по км до аэропорта
def mean_corr(data,column, alpha=0.5):
    mean_data = data.copy()
    mean_data['km_center'] = (mean_data[column]/1000)
    mean_data['km_center'] = mean_data['km_center'].astype(int)

    mean_data = (mean_data.groupby('km_center')
        .agg({'price_m2':np.mean}))

    mean_data['km'] = mean_data.index
    display_corr(mean_data,'km','price_m2', alpha=0.5)
```
